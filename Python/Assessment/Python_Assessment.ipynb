```python
data = {}


print("Welcome to Fruit Market!!")
print("1. Store Manager")
print("2. Customer")

role = int(input("Enter your role (1 or 2): "))
if role == 1:
    while True:
        print("\nStore Manager Menu:")
        print("1. Add Fruit")
        print("2. View Inventory")
        print("3. Update Fruit Stock")
        print("4. Exit")
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            fruit = input("Enter fruit name: ")
            price = float(input("Enter fruit price: "))
            stock = int(input("Enter fruit stock: "))
            data[fruit] = {'price': price, 'stock': stock}
            print(f"{fruit} added to inventory.")
        
        elif choice == 2:
            if not data:
                print("Inventory is empty.")
            else:
                print("\nCurrent Inventory:")
                for fruit, details in data.items():
                    print(f"{fruit}: Price - ${details['price']}, Stock - {details['stock']}")
        
        elif choice == 3:
            fruit = input("Enter fruit name to update: ")
            if fruit in data:
                stock = int(input("Enter new stock quantity: "))
                data[fruit]['stock'] = stock
                print(f"{fruit} stock updated to {stock}.")
            else:
                print(f"{fruit} not found in inventory.")
                
        elif choice == 4:
            print("Exiting Store Manager Menu.")
            break
        
        else:
            print("Invalid choice. Please try again.")
            
elif role == 2:
    while True:
        print("\nCustomer Menu:")
        print("1. View Available Fruits")
        print("2. Purchase Fruit")
        print("3. Exit")
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            if not data:
                print("No fruits available.")
            else:
                print("\nAvailable Fruits:")
                for fruit, details in data.items():
                    if details['stock'] > 0:
                        print(f"{fruit}: Price - ${details['price']}, Stock - {details['stock']}")
        
        elif choice == 2:
            fruit = input("Enter fruit name to purchase: ")
            if fruit in data and data[fruit]['stock'] > 0:
                quantity = int(input("Enter quantity to purchase: "))
                if quantity <= data[fruit]['stock']:
                    total_cost = quantity * data[fruit]['price']
                    data[fruit]['stock'] -= quantity
                    print(f"Purchased {quantity} of {fruit}. Total cost: ${total_cost}")
                else:
                    print(f"Only {data[fruit]['stock']} of {fruit} available.")
            else:
                print(f"{fruit} not available or out of stock.")
                
        elif choice == 3:
            print("Exiting Customer Menu.")
            break
        
        else:
            print("Invalid choice. Please try again.")

else:
    print("Invalid role selected. Please restart the program.")     
```
